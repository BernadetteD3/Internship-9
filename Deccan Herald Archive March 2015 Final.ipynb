# Deccan Herald Archives for the month of March 2015

```python
# Importing Libraries
import pandas as pd
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import ElementNotVisibleException
import time
from selenium.common.exceptions import SessionNotCreatedException
from selenium.common.exceptions import TimeoutException
```

```python
driver = webdriver.Firefox(executable_path='C:\GeckoDriver\geckodriver.exe')
```

```python
url ='https://www.deccanherald.com/archives'
```

```python
# Open the webpage through the web driver
driver.get(url)
```

```python
INTENTIONAL ERROR: You have to click the " I'll do this later " button on the browser manually before running the rest.
```

```python
Date_list=[]
Author_list=[]
Vertical_list=[]
Headline_list=[]
Description_list=[]

for i in range(0,32):
    
    # Open the webpage through the web driver
    driver.get(url)
    
    # Wait for the page to update 
    time.sleep(5)
    
    # clicking the calendar button
    calender_button=driver.find_element_by_xpath("/html/body/main/div[3]/div/div/div/div/div/div/div/div/div/section/div/div/div[2]/div[1]/div[2]/form/div/div/a/img")
    calender_button.click()
    
    # navigating to the start date via years
    for x in range(6):
        down_button=driver.find_element_by_xpath("/html/body/div[2]/div[1]/div/div/span[2]")
        down_button.click()
    
    #  navigating to the start date via months
    for y in range(3):
        right_button=driver.find_element_by_xpath("//span[@class='flatpickr-next-month']")
        right_button.click()
        
    # get the link to the element 
    date_button = driver.find_elements_by_class_name("flatpickr-day ")[i]
    
    # click on the day button.
    if date_button.get_attribute('class') == "flatpickr-day ":
        date_button.click()
        
        # clicking submit button
        submit_button=driver.find_element_by_id("getArchive")
        submit_button.click()
        
        # Wait for the page to update 
        time.sleep(5)
        
        links=driver.find_elements_by_xpath('/html/body/main/div[3]/div/div/div/div/div/div/div/div/div/section/div/div/div[3]/div[1]/div[*]/ul/li[*]/a' and '/html/body/main/div[3]/div/div/div/div/div/div/div/div/div/section/div/div/div[3]/div[1]/div[*]/ul/li/a')      
        links_list = []
        for link in links:
            print(link.get_attribute('href'))
            # Fetch and store the links for this date
            links_list.append(link.get_attribute('href'))

        
        # Loop through all the links for this date and launch one by one

        for lin in links_list[0:5]:
    
            driver.get(lin)
            time.sleep(5)  
    
            dates=driver.find_element_by_xpath('/html/body/main/div[4]/div/div/div[3]/div[2]/ul/li[1]')
            authors=driver.find_element_by_xpath('/html/body/main/div[4]/div/div/div[3]/div[2]/span' )
            vertices=driver.find_element_by_xpath('/html/body/main/div[1]/div/div/ul/li[2]/a')
            headlines=driver.find_element_by_xpath('/html/body/main/div[1]/div/div/ul/li[3]')
            descriptions=driver.find_element_by_xpath('/html/body/main/div[4]/div/div/div[4]/div/div[2]/div[2]/div[*]/div/div/p[1]' or '/html/body/main/div[4]/div/div/div[4]/div/div[2]/div[2]/div/div/div/p[1]')
                                                       
  
            date=dates.text
            Date_list.append(date)
     
            author=authors.text
            Author_list.append(author)
    
            verticle=vertices.text
            Vertical_list.append(verticle)
    
            headline=headlines.text
            Headline_list.append(headline)
    
            
            try:
                description=descriptions.text
                Description_list.append(description)
    
            except NoSuchElementException as e:
                print("Exception Raised : ", e)                    # Printing Message
                new_desc= driver.find_element_by_xpath('//div[@align="justify"]')        # Replacing previous element by a valid element which appears on webpage
                description = new_desc.text
                Description_list.append(description)
   
    

March_2015_News=pd.DataFrame({'Date':Date_list,
                                'Author':Author_list,
                                'Verticle':Vertical_list,
                                'Headline':Headline_list,
                                'Description':Description_list})
print(March_2015_News)      
```
